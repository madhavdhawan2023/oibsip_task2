#  Analyzing Data By barplots

```python
fg = px.bar(df,x='Region',y='Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame='Date',template='plotly', height=600) 
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#  Analyzing Data By Funnel Chart

```python
fg = px.funnel(df,x='Region',y='Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame='Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

# Analyzing Data By Boxplot

```python
fg = px.box(df,x='Region',y='Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame='Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#  Analyzing Data By scatterplot

```python
fg = px.scatter(df,x='Region',y='Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame='Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#  Analyzing Data By Histogram

```python
fg = px.histogram(df,x='Estimated Unemployment Rate (%)',y='Region',color='Region',title='Unemployment rate',animation_frame='Date',template='plotly',height=1100)
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#  Analyzing Data By Pie Chart

```python

d = df.Area.value_counts().to_dict()

fg = plt.figure(figsize = (18, 6))
ax = fg.add_subplot()

cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 6, 24)]

ax.pie(d.values(), 
       labels = d.keys(), 
       autopct = '%1.1f%%', 
       textprops = {'fontsize': 10, 'color' : "white"} ,
       colors= colors
      )


ax.set_title("\n\nUnemployment in Urban and Rural Areas")


ax.legend(loc = "upper left", bbox_to_anchor = (1, 0, 0.5, 1), fontsize =10, title = "Areas");
```

#  Analyzing Data By Line Graph

```python
fg = px.line(df,y='Estimated Unemployment Rate (%)',x='Region',title='Unemployment rate',animation_frame='Date',template='plotly', height=700)
fg.update_layout(yaxis={'categoryorder':'total descending'})
fg.show()
```
